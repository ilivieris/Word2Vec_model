```python
from gensim import utils
import json
from tqdm import tqdm
import re 

L = []
# iterate over the plain text data we just created
with utils.open('./Data/elwiki-latest.json.gz', 'rb') as f:
    loop = tqdm(f, leave=True)
    for line in loop:
        # decode each JSON line into a Python dictionary object
        article = json.loads(line)
        # print()
        # each article has a "title", a mapping of interlinks and a list of "section_titles" and
        # "section_texts".
        # print("Article title: %s" % article['title'])
        # print("Interlinks: ", article['interlinks'])
        # for section_title, section_text in zip(article['section_titles'], article['section_texts']):
        #     print("Section title: %s" % section_title)
        #     print("Section text: %s" % section_text)
        

        for text in article['section_texts']:
            text = text.replace("'", "").replace("*", "").replace("\n", ".")
            text = re.sub(r'\.+', ".", text)
            
            while len(text) > 1:
                if len(re.findall("^[A-Z]", text) + re.findall("^[a-z]", text) + re.findall("^[Α-Ω]", text) + re.findall("^[α-ω]", text)) >= 1: break
                text = text[1:]
                
            if len(text) > 10: L.append(text)

        loop.set_description(f"{article['title']} - [{len(L)}]")
        
        
        if len(L) > 4: break
        # if 'Μιχάλης Κακογιάννης' == article['title']: break
        # print(len(text))
        # print( len(article['section_texts']) )
```
